```python
import pandas as pd

df = pd.read_csv("mail_data (2).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
mail_data = df.where(pd.notnull(df), '')
```

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

```python
X = mail_data['Message']
Y = mail_data['Category'].astype(int)
```

```python
X = df['Message']
Y = df['Category']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=3
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

feature_extraction = TfidfVectorizer(
    min_df=1,
    stop_words='english',
    lowercase=True
)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
Y_train = Y_train.map({'ham': 1, 'spam': 0})
Y_test = Y_test.map({'ham': 1, 'spam': 0})
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X_train_features, Y_train)
```

```python
from sklearn.metrics import accuracy_score

train_prediction = model.predict(X_train_features)

train_accuracy = accuracy_score(Y_train, train_prediction)

print("Training Accuracy:", train_accuracy)
```

```python
test_prediction = model.predict(X_test_features)

test_accuracy = accuracy_score(Y_test, test_prediction)

print("Testing Accuracy:", test_accuracy)
```

```python
input_mail = ["Congratulations! You have won a free iPhone. Click here to claim your prize."]

input_data_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_data_features)

if prediction[0] == 1:
    print("Ham Mail")
else:
    print("Spam Mail")
```

```python
input_mail = ["Hey! Are we still meeting at 5 PM today?"]

input_data_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_data_features)

if prediction[0] == 1:
    print("Ham Mail")
else:
    print("Spam Mail")
```

```python
input_mail = ["Congratulations! You have won a $1000 gift card. Click here to claim your prize now!"]

input_data_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_data_features)

if prediction[0] == 1:
    print("Ham Mail")
else:
    print("Spam Mail")
```

```python
import pickle

pickle.dump(model, open("spam_model.pkl", "wb"))
pickle.dump(feature_extraction, open("vectorizer.pkl", "wb"))

print("Model Saved Successfully!")
```

```python
%%writefile app.py

import streamlit as st
import pickle

# Load model and vectorizer
model = pickle.load(open("spam_model.pkl", "rb"))
vectorizer = pickle.load(open("vectorizer.pkl", "rb"))

st.title("📧 Spam Mail Detection")
st.write("Enter an email or SMS message to check if it is Spam or Ham.")

message = st.text_area("Enter your message")

if st.button("Predict"):
    message_vector = vectorizer.transform([message])
    prediction = model.predict(message_vector)

    if prediction[0] == 1:
        st.success("✅ Ham Mail")
    else:
        st.error("🚫 Spam Mail")
```
